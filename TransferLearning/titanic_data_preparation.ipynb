# Pré-traitement de la dataset du titanic

```python
import pandas as pd

# Importer les données train et test de la dataset
df_train = pd.read_csv('./titanic/train.csv')
df_test = pd.read_csv('./titanic/test.csv')
df_train.head()
```

```python
df_test.head()
```

```python
# Séléctionner les features les plus utiles pour le train: X_train
X_train = df_train[["Pclass",
"Sex",
"Age",
"SibSp",
"Parch",
"Fare",
"Embarked",
"Survived"]]
X_train.head()
```

```python
# Séléctionner les features les plus utiles pour le test: X_test
X_test = df_test[["Pclass",
"Sex",
"Age",
"SibSp",
"Parch",
"Fare",
"Embarked"]]
X_test.head()
```

```python
# Y_test
df_y_test = pd.read_csv("./titanic/gender_submission.csv")
y_test = df_y_test[["Survived"]]
X_test["Survived"] = y_test.values
X_test.head()
```

## Nettoyer la dataset: supprimer/remplacer les NaN

```python
print("Le nombre de valeurs nulles dans chaque colonne du train: \n {}".format(X_train.isnull().sum()))
print("Le nombre de valeurs nulles dans chaque colonne du test: \n {}".format(X_test.isnull().sum()))
```

```python
# Supprimer les exemples ayant une valeur nulle sur la feature 'embarked'
X_train = X_train[X_train["Embarked"].notnull()]
X_train.isnull().sum()
```

```python
# Supprimer les exemples ayant une valeur nulle sur la feature 'Fare'
X_test = X_test[X_test["Fare"].notnull()]
X_test.isnull().sum()
```

```python
# Remplacer les valeurs nulles par la moyenne
X_train["Age"] = X_train["Age"].fillna(X_train["Age"].mean())
X_test["Age"] = X_test["Age"].fillna(X_test["Age"].mean())
X_train[26:36]
```

## Transformer les valeurs qualitatives en valeurs qualitatives

```python
# Train dataset
X_train = pd.get_dummies(X_train, prefix=['Sex', 'Embarked'], columns=['Sex','Embarked'])
X_train.head()
```

```python
# Test dataset
X_test = pd.get_dummies(X_test, prefix=['Sex', 'Embarked'], columns=['Sex','Embarked'])
X_test.head()
```

## Normaliser les données

```python
from sklearn.preprocessing import MinMaxScaler

# Normalisation Min Max de la dataset train
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_train = pd.DataFrame(data=X_train_scaled, columns=X_train.columns)
X_train.head()
```

```python
# Normalisation Min Max de la dataset test
scaler = MinMaxScaler()
X_test_scaled = scaler.fit_transform(X_test)
X_test = pd.DataFrame(data=X_test_scaled, columns=X_test.columns)
X_test.head()
```

```python
# Y_train
y_train = X_train[["Survived"]]
X_train.drop(['Survived'],axis=1, inplace=True)
y_train.head()
```

```python
# Y_test
y_test_column = X_test['Survived']
y_test = pd.DataFrame(data=y_test_column, columns=["Survived"])
X_test.drop(['Survived'],axis=1, inplace=True)
y_test.head()
```
